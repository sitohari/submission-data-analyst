## Exploratory Data Analysis (EDA)

### Explore `day_csv` 

```python
day_df.describe(include="all")
```

```python
daily_counts = day_df.groupby('weekday')['cnt'].sum()
```

```python
day_df["dteday"] = pd.to_datetime(day_df["dteday"])
day_df["year"] = day_df["dteday"].dt.year
```

```python
yearly_counts_day = day_df.groupby('year')['cnt'].sum()
```

### Explore `hour_csv` 

```python
hour_df.describe(include="all")
```

## Visualization & Explanatory Analysis

### 1. Hari apa dalam seminggu yang paling banyak sepeda disewa?


```python
sns.barplot(x=daily_counts.index, y=daily_counts.values)
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Sepeda Disewa')
plt.title('Jumlah Penyewaan Sepeda per Hari')
plt.show()
```

### 2. Apakah jumlah pengguna sepeda meningkat dari tahun ke tahun?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_counts_day.index, y=yearly_counts_day.values)
plt.xlabel('Tahun')
plt.ylabel('Total Penyewaan Sepeda')
plt.title('Perkembangan Jumlah Penyewaan Sepeda per Tahun')
plt.show()
```

## RFM Analysis

```python
daily_counts = day_df.groupby('weekday')['cnt'].sum()
sns.barplot(x=daily_counts.index, y=daily_counts.values)
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Sepeda Disewa')
plt.title('Jumlah Penyewaan Sepeda per Hari')
plt.show()
```

```python
yearly_counts = day_df.groupby('year')['cnt'].sum()
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values)
plt.xlabel('Tahun')
plt.ylabel('Total Penyewaan Sepeda')
plt.title('Perkembangan Jumlah Penyewaan Sepeda per Tahun')
plt.show()
```
